```
def question_five():
  df2 = df.copy()
  df2['sepal_length'] = df2['sepal_length'].apply(fn)
  return df2
  
def fn(a):
    a = a * 10
    return int(a/10) + (a%10)
  
question_five()
```

Question 6

```
def question_six():
  df2 = df.copy()
  s = df2['species'].unique()
  variation = 0
  for i in s:
    df3 = df2[df2['species'] == i]
    maxi = df3['sepal_width'].max()
    mini = df3['sepal_width'].min()
    temp = maxi - mini
    if temp > variation:
      variation  = maxi - mini
      species = i
      
  return species

question_six()
```

Question 7

```
def question_seven():
  df2 = df.copy()
  df2 = df2[df2['species'] == 'versicolor']
  avg_length = df2['sepal_length'].mean()
  
  df2 = df.copy()
  df2 =  df2[df2['species'] == 'virginica']
  count = df2[df2['sepal_length'] > avg_length]['species'].count()
  return count

question_seven()
```
